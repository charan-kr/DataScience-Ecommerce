```python
def flatten_json(y):
    out = {}
    def flatten(x, name=""):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name+a+"_")
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name+str(i)+"_")
                i+=1
        else:
            out[name[:-1]] = x
    
    flatten(y)
    
    return out
```

```python
normalized_data = [flatten_json(product) for product in data]
```

```python
df = pd.DataFrame(normalized_data)
```
